```python
import numpy as np
import pandas as pd
import matplotlib as pyplot
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
independent
```

```python
dependent=dataset[['classification_yes']]
```

```python
dataset["classification_yes"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.03,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB 
param_grid={'var_smoothing':[1e-9,1e-8,1e-7]}
grid=GaussianNB()
grid_search=GridSearchCV(grid,param_grid,cv=5)
grid_search.fit(x_train,y_train)
```

```python
grid_pred=grid_search.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_pred)
cm
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_pred)
print(clf_report)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import BernoulliNB 
param_grid={'alpha':[1.0,2.0,3.0],"force_alpha":[True,False]}
grid=BernoulliNB()
grid_search=GridSearchCV(grid,param_grid,cv=5)
grid_search.fit(x_train,y_train)
grid_pred=grid_search.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_pred)
cm
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_pred)
print(clf_report)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import MultinomialNB
param_grid={"force_alpha":[True]}
grid=MultinomialNB()
grid_search=GridSearchCV(grid,param_grid,cv=10)
grid_search.fit(x_train,y_train)
grid_pred=grid_search.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_pred)
cm
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_pred)
print(clf_report)
```
